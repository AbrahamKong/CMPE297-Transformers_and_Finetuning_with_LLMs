```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained(
        "jinaai/starcoder-1b-textbook", device_map='auto'
    )

tokenizer = AutoTokenizer.from_pretrained("jinaai/starcoder-1b-textbook")
```

## Prompt Example for Python Generation

```python
prompt = '''
generate a python function that given a positive integer, it returns its roman numeral equivalent as a string,

def int_to_mini_roman
'''

inputs = tokenizer(prompt.rstrip(), return_tensors="pt").to("cuda")

generation_output = model.generate(
    **inputs,
    max_new_tokens=256,
    eos_token_id=tokenizer.eos_token_id,
    return_dict_in_generate=True,
)

s = generation_output.sequences[0]
output = tokenizer.decode(s, skip_special_tokens=True)

print(output)
```

My Prompts

```python
def transform (prompt):

  inputs = tokenizer(prompt.rstrip(), return_tensors="pt").to("cuda")

  generation_output = model.generate(
      **inputs,
      max_new_tokens=256,
      eos_token_id=tokenizer.eos_token_id,
      return_dict_in_generate=True,
  )

  s = generation_output.sequences[0]
  output = tokenizer.decode(s, skip_special_tokens=True)

  print(output)
```

```python
prompt = '''Translate the following Python code to Java:

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
'''
transform(prompt)
```

```python
prompt = '''Explain every line of code and write it in comment of this code

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
'''
transform(prompt)
```

```python
prompt = ''' Generate a a data set of random animal names

and generate a function that uses the link list data structure
and use it to iterate through the dataset

Show the actual code
'''
transform(prompt)
```
